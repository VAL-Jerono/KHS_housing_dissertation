```python

# Cell 3 — Run conversion
import sys
sys.path.insert(0, 'src')
import pandas as pd
import os
from config import DTA_FILES, RAW_DATA, PARQUET

os.makedirs(PARQUET, exist_ok=True)

for key, filename in DTA_FILES.items():
    dta_path = RAW_DATA / filename
    parquet_path = PARQUET / filename.replace('.dta', '.parquet')

    if not dta_path.exists():
        print(f"⚠  {key}: file not found, skipping")
        continue
    if parquet_path.exists():
        print(f"✓  {key}: already converted")
        continue

    print(f"→  Converting {key}...")
    df = pd.read_stata(dta_path, convert_categoricals=False)
    df.to_parquet(parquet_path, index=False)
    size_mb = os.path.getsize(parquet_path) / 1e6
    print(f"   Saved: {parquet_path.name} ({size_mb:.2f} MB)")

print("\nAll done.")
```
